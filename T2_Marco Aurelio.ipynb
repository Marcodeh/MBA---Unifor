```r
# Carregando os pacotes essenciais
library(tidyverse)    # Inclui dplyr, ggplot2 e outros
```

```r
# Função para baixar arquivos do Google Drive
down_gdrive <- function(link){

    # Quebra o caractere pela barra "/", resultando numa lista de caracteres
    id <- strsplit(link, split = "/")
    # Acessa o primeiro elemento da lista e posteriormente o sexto elemento.
    id <- id[[1]][6]
    # Download do arquivo
    system(paste("gdown --id ",id))
}
```

```r
# Baixa o arquivo do Google Drive a partir do link compartilhado
link <- 'https://drive.google.com/file/d/1DNYmJbHIFqX7PxOe5XMH6sizfrvQwIuy/view?usp=sharing'
down_gdrive(link)
```

```r
# Carrega o arquivos 'csv' separados por ',' e variáveis (colunas) do tipo "character"
nba <- read.csv(
    '/content/nba_stats.csv',
    sep=',',
    colClasses = 'character'
)
head(nba)
```

```r
# Retorna os primeiros elementos de um vetor, matriz, tabela ou dataframe.
str(nba)
```

```r
# Visualizando a estrutura dos dados
glimpse(nba)
```

```r
# Verificando o nome correto da coluna de salário
names(nba)  # Se precisar confirmar o nome da coluna original
```

```r
# Convertendo a coluna de salário de texto para número
nba$salary_numeric <- nba$salary |>
  str_replace_all("[$,]", "") |>
  as.numeric()
```

```r
# Conferindo se a conversão funcionou
summary(nba$salary_numeric)
```

```r
# Checando quantidade de valores faltantes por coluna
colSums(is.na(nba))
```

```r
# Criando um novo dataframe com colunas úteis
nba_selected <- nba |>
  select(name, year, teamId, positionId, avgPoints, avgAssists, avgRebounds, avgMinutes, salary_numeric)
```

```r
# Visualizando os dados filtrados
head(nba_selected)
```

```r
# Gráfico de distribuição dos salários
ggplot(nba_selected, aes(x = salary_numeric)) +
  geom_histogram(bins = 30, fill = "steelblue", color = "black") +
  labs(
    title = "Distribuição dos Salários dos Jogadores",
    x = "Salário (USD)",
    y = "Quantidade de Jogadores"
  ) +
  scale_x_continuous(labels = scales::dollar)
```

```r
# Gráfico de boxplot dos salários por posição
ggplot(nba_selected, aes(x = positionId, y = salary_numeric)) +
  geom_boxplot(fill = "orange", color = "black") +
  labs(
    title = "Salário por Posição",
    x = "Posição",
    y = "Salário (USD)"
  ) +
  scale_y_continuous(labels = scales::dollar)
```

```r
ggplot(nba_selected, aes(x = avgPoints, y = salary_numeric)) +
  geom_point(color = "darkgreen", alpha = 0.6) +  # pontos da dispersão
  geom_smooth(method = "lm", se = FALSE, color = "black") +  # linha de tendência
  labs(
    title = "Relação entre Pontos por Jogo e Salário",
    x = "Pontos por Jogo (avg_points)",
    y = "Salário (USD)"
  ) +
  scale_y_continuous(labels = scales::dollar)
```

```r
ggplot(nba_selected, aes(x = avgAssists, y = salary_numeric)) +
  geom_point(color = "purple", alpha = 0.6) +
  geom_smooth(method = "lm", se = FALSE, color = "black") +
  labs(
    title = "Relação entre Assistências por Jogo e Salário",
    x = "Assistências por Jogo (avg_assists)",
    y = "Salário (USD)"
  ) +
  scale_y_continuous(labels = scales::dollar)
```

```r
ggplot(nba_selected, aes(x = avgRebounds, y = salary_numeric)) +
  geom_point(color = "blue", alpha = 0.6) +
  geom_smooth(method = "lm", se = FALSE, color = "black") +
  labs(
    title = "Relação entre Rebotes por Jogo e Salário",
    x = "Rebotes por Jogo (avg_rebounds)",
    y = "Salário (USD)"
  ) +
  scale_y_continuous(labels = scales::dollar)
```

```r
# Selecionando apenas colunas numéricas relevantes (sem NAs)
nba_cor <- nba_selected |>
  select(avgPoints, avgAssists, avgRebounds, avgMinutes, salary_numeric) |>
  na.omit()
```

```r
# Conferindo os tipos das colunas
str(nba_cor)
```

```r
# Garantindo que todas as colunas sejam numéricas
nba_cor <- nba_cor |>
  mutate(across(everything(), as.numeric))
```

```r
# Calculando matriz de correlação
cor_matrix <- cor(nba_cor)
```

```r
# Exibindo a matriz
print(round(cor_matrix, 2))
```
